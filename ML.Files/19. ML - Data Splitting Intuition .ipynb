# Forword_propagation 

```python
def forword_propagation(X_train , parameters):
    m = parameters['m']
    c = parameters['c']
    
    predection = np.multiply(X_train , parameters) + c
    
    return predection
```

# Cost_function 

```python
def cost_function(predection , Y_train):
    
    cost = np.mean((predection - Y_train)**2) * 0.5      # 1/2 = 0.5
    
    return cost
```

# Backword_propagation 

```python
def backword_propagation(X_train , Y_train , predection) :
    
    derivatives = dict()
    
    df = predection - Y_test
    dm = np.mean(np.multiply(df , X_train))
    dc = np.mean(df)
    
    dm = derivatives['dm']
    dc = derivatives['dc']
    
    return derivatives
```

# Update the parameters 

```python
def update_parameters(parameters , derivatives , learning_rate):
    parameters['m'] = parameters['m'] - learning_rate * derivatives['dm']
    parameters['c'] = parameters['m'] - learning_rate * derivatives['dc']
    
    return parameters
```
